```python
class mobil:
  def __init__(self,merk, model, tahun):
    self.merk = merk
    self.model = model
    self.tahun = tahun

def main():
  Mobil = mobil("Toyota", "Avanza", 2020)

  print("Merk = ", Mobil.merk)
  print("Model = ", Mobil.model)
  print("Tahun = ", Mobil.tahun)

if __name__ == "__main__":
  main()
```
