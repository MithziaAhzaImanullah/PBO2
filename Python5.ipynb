```python
class laptop:
  def __init__(self, brand, type, harga):
    self.brand = brand
    self.type = type
    self.harga = harga

def main():
  laptop1 = laptop("Asus", "Vivobook", 6000000)
  laptop2 = laptop("HP", "Victus 16", 15000000)

  print(laptop1.brand, laptop1.type, laptop1.harga)
  print(laptop2.brand, laptop2.type, laptop2.harga)

if __name__ == "__main__":
  main()
```
